```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

```

```python
# Cargar la imagen en escala de grises
image = cv2.imread('escalera.png', cv2.IMREAD_GRAYSCALE)
plt.imshow(image, cmap='gray')
plt.show()
```

```python
# Aplicar el algoritmo de Canny para la detección de bordes
bordes = cv2.Canny(image, 50, 150, apertureSize=3)
plt.imshow(bordes, cmap='gray')
plt.show()
```

```python
# Aplicar la Transformada de Hough para detectar líneas
lineas = cv2.HoughLines(bordes, 1, np.pi / 180, 200)

# Dibujar las líneas detectadas en la imagen original
if lineas is not None:
    for p, theta in lines[:,0]:
        a = np.cos(theta)
        b = np.sin(theta)
        x0 = a * p
        y0 = b * p
        x1 = int(x0 + 1000 * (-b))
        y1 = int(y0 + 1000 * (a))
        x2 = int(x0 - 1000 * (-b))
        y2 = int(y0 - 1000 * (a))
        cv2.line(image, (x1, y1), (x2, y2), (255, 255, 0), 2)

# Mostrar la imagen original con las líneas detectadas
plt.imshow(image, cmap='gray')
plt.show()
```
